```python
import os
import cv2
import pandas as pd
from PIL import Image, ImageDraw, ImageColor

filePath = "./res/lab6.xlsx"
imagesFolder = "./res/Images"
outputFolder = "./res/output/"
```

```python
def modify_glasses(df, n, glasses_color=None, thickness=None):
    if glasses_color is None: glasses_color = df.at[n, 'glasses color']
    if thickness is None: thickness = df.at[n, 'line width']
        
    df.at[n, 'glasses color'] = glasses_color
    df.at[n, 'line width'] = thickness
    
    df = df.loc[[n]].reset_index(drop=True)
    return df
```

```python
CASCADE_FACE = cv2.data.haarcascades + "haarcascade_frontalface_default.xml"
CASCADE_EYES = cv2.data.haarcascades + "haarcascade_eye.xml"

face_cascade = cv2.CascadeClassifier(CASCADE_FACE)
eye_cascade = cv2.CascadeClassifier(CASCADE_EYES)

def add_glasses_to_image(img_df, images_folder, output_folder):
    img_file_name = img_df["file name"]
    glasses_color = img_df["glasses color"]
    thickness = int(img_df["line width"])
    outline = img_df["glasses color"]
    image_size = img_df["image size"]
    print(img_file_name)
    
    img_path = os.path.join(images_folder, img_file_name)
    if not os.path.exists(img_path):
        print(f"Файл не знайдено: {img_path}")
        return

    try:
        image_size = image_size.split('x')
        width = int(image_size[0])
        height = int(image_size[1])
    except ValueError:
        print(f"Некоректний розмір зображення: {image_size}")
        return
        
    img = cv2.imread(img_path)
    original_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    original_img = Image.fromarray(original_rgb, mode="RGB")

    img_resized = cv2.resize(img, (width, height), interpolation=cv2.INTER_AREA)
    gray = cv2.cvtColor(img_resized, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray, 1.2)
    eyes = eye_cascade.detectMultiScale(gray, 1.2)
    rgb = cv2.cvtColor(img_resized, cv2.COLOR_BGR2RGB)
    pil_img = Image.fromarray(rgb, mode="RGB")
    draw = ImageDraw.Draw(pil_img)
    print(eyes)
    eye_coords = sorted(eyes, key=lambda x: x[0])
    (ex1, ey1, ew1, eh1) = eye_coords[0]
    (ex2, ey2, ew2, eh2) = eye_coords[1]
    center1 = (ex1 + ew1 // 2, ey1 + eh1 // 2)
    center2 = (ex2 + ew2 // 2, ey2 + eh2 // 2)
    radius1 = max(ew1, eh1) // 2
    radius2 = max(ew2, eh2) // 2
    
    for x, y, w, h in faces:
        draw.rectangle((x, y, x+w, y+h), outline=outline)

        for (ex, ey, ew, eh) in eyes:
            center = (ex + ew // 2, ey + eh // 2)
            radius = max(ew, eh) // 2
            draw.ellipse(
                (center[0] - radius, center[1] - radius, center[0] + radius, center[1] + radius),
                outline = outline,
                width= thickness
            )

    bridge_start = (center1[0] + radius1, center1[1])
    bridge_end = (center2[0] - radius2, center2[1])
    draw.line([bridge_start, bridge_end], fill=outline, width=thickness)

    left_bracket_start = (center1[0] - radius1, center1[1]) 
    left_bracket_end = (center1[0] - 3 * radius1, center1[1] + 1 * radius1)  
    right_bracket_start = (center2[0] + radius2, center2[1]) 
    right_bracket_end = (center2[0] + 3 * radius2, center2[1] + 1 * radius2)  
    draw.line([left_bracket_start, left_bracket_end], fill=outline, width=thickness)
    draw.line([right_bracket_start, right_bracket_end], fill=outline, width=thickness)
    
    output_path = os.path.join(output_folder, f"processed_{img_file_name}")
    pil_img.save(output_path)

    display(original_img)
    display(pil_img)
```

```python
os.makedirs(outputFolder, exist_ok=True)
df = pd.read_excel(filePath)
n = 2
if (n == 2): df.at[n, 'glasses color'] = "cyan"
    
modified_df = modify_glasses(df=df, n=n)
add_glasses_to_image(modified_df.loc[0], imagesFolder, outputFolder)

modified_df = modify_glasses(df=df, n=0)
add_glasses_to_image(modified_df.loc[0], imagesFolder, outputFolder)

modified_df = modify_glasses(df=df, n=1, glasses_color='Green', thickness=1)
add_glasses_to_image(modified_df.loc[0], imagesFolder, outputFolder)

modified_df = modify_glasses(df=df, n=3)
add_glasses_to_image(modified_df.loc[0], imagesFolder, outputFolder)

modified_df = modify_glasses(df=df, n=4)
add_glasses_to_image(modified_df.loc[0], imagesFolder, outputFolder)
```

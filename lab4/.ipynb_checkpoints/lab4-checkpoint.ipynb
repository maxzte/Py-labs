```python
import os
import glob
import pandas as pd
import base64
import random
from PIL import Image, ImageEnhance
from io import BytesIO
from IPython.display import HTML

folderPath = "./res"
```

```python
def analyze_images(folder_path):
    image_files = glob.glob(os.path.join(folder_path, "*.*"))
    data = []

    for file in image_files:
        try:
            with Image.open(file) as img:
                file_name = os.path.basename(file)
                color_model = img.mode  # Кольорова модель (наприклад, RGB, L)
                img_format = img.format  # Формат файлу (наприклад, JPEG, PNG)
                num_channels = len(img.getbands())  # Кількість каналів
                file_size_mb = round(os.path.getsize(file) / (1024 * 1024), 2)  # Розмір файлу в MB
                width, height = img.size  # Розмір зображення
                file_path = os.path.abspath(file)  # Повний шлях до файлу
                resized_image = img.resize((200, 200))  # Зменшене зображення (100x100)

                image_html = image_formatter(resized_image)
                
                data.append({
                    "File Name": file_name,
                    "Color Model": color_model,
                    "Format": img_format,
                    "Number of Channels": num_channels,
                    "File Size (MB)": file_size_mb,
                    "Width": width,
                    "Height": height,
                    "Full Path": file_path,
                    "Resized Image": image_html
                })
        except Exception as e:
            print(f"Error processing file {file}: {e}")

    df = pd.DataFrame(data)
    return df
```

```python
def image_formatter(im):
    buffered = BytesIO()
    im.save(buffered, format="PNG")
    img_str = base64.b64encode(buffered.getvalue()).decode()
    return f'<img src="data:image/png;base64,{img_str}" style="display:block; margin:auto;"/>'
```

```python
df_images = analyze_images(folderPath)

html_table = df_images.to_html(
    escape=False,
    formatters={"Resized Image": lambda x: x}
)

display(HTML(html_table))
```

```python
def create_poster(folder_path, output_name, rows, cols):
    supported_formats = (".jpg", ".jpeg", ".png", ".bmp", ".gif", ".tiff", ".webp")
    image_files = [f for f in glob.glob(os.path.join(folder_path, "*.*")) if f.lower().endswith(supported_formats)]
    
    if not image_files:
        raise ValueError("У папці немає підтримуваних зображень.")
    
    selected_images = [random.choice(image_files) for _ in range(rows * cols)]
    thumbnail_size = (250, 250)
    processed_images = []
    for img_path in selected_images:
        try:
            with Image.open(img_path) as img:
                img = crop_to_square(img)
                img = img.resize(thumbnail_size)
                img, _ = apply_random_filter(img) 
                processed_images.append(img)
        except Exception as e:
            print(f"Пропущено файл {img_path}: {e}")
    
    if len(processed_images) < rows * cols:
        raise ValueError("Недостатньо зображень для створення постера.")
    
    # Створення полотна постера
    poster_width = cols * thumbnail_size[0]
    poster_height = rows * thumbnail_size[1]
    poster = Image.new("RGB", (poster_width, poster_height), "white")
    
    # Додавання зображень на постер
    for i in range(rows):
        for j in range(cols):
            x = j * thumbnail_size[0]
            y = i * thumbnail_size[1]
            poster.paste(processed_images[i * cols + j], (x, y))
    
    output_path = os.path.join(folder_path, f"{output_name}.jpg")
    poster.save(output_path)
    display(poster)
```

```python
def apply_random_filter(image):
    filters = {
        "red": (1, 0.5, 0.5),
        "green": (0.5, 1, 0.5),
        "blue": (0.5, 0.5, 1),
        "yellow": (1, 1, 0.5),
        "magenta": (1, 0.5, 1),
        "cyan": (0.5, 1, 1)
    }
    filter_choice = random.choice(list(filters.keys()))
    r, g, b = filters[filter_choice]
    return ImageEnhance.Color(image).enhance(r), filter_choice
```

```python
def crop_to_square(image):
    width, height = image.size
    new_edge = min(width, height)
    left = (width - new_edge) // 2
    top = (height - new_edge) // 2
    right = left + new_edge
    bottom = top + new_edge
    return image.crop((left, top, right, bottom))
```

```python
create_poster(folderPath, "poster", 2, 2)
```
